# Linear Regression with one variable
## 1. Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Load training dataset

```python
DATA_FILE_NAME = '../../dataset/1_linearinput.csv'

# load data
df = pd.read_csv(DATA_FILE_NAME)
# plot data
df.plot(x='x', y='y', legend=False, marker='o', style='o', mec='b', mfc='w')
# expected line
plt.plot(df.values[:,0], df.values[:,1], color='g')
plt.xlabel('x'); plt.ylabel('y'); plt.show()

# extract X, y
X = df.values[:,0]
y = df.values[:,2]
m = y.size # number of training examples
# add X_0 to X
X = np.concatenate((np.ones((m,1)), X.reshape(-1,1)), axis=1)
```

## 3. Learing parameters

```python
W = np.dot(np.linalg.pinv(np.dot(X.T, X)), np.dot(X.T, y))
print 'W :=', W
```

## 4. Plot result

```python
# training data
df.plot(x='x', y='y', legend=False, marker='o', style='o', mec='b', mfc='w')
# expected line
plt.plot(df.values[:,0], df.values[:,1], color='g')
# predicted line
y_hat = np.dot(X, W)
plt.plot(df.values[:,0], y_hat, color='r')
plt.xlabel('x'); plt.ylabel('y'); plt.show()
```
